```python
from dotenv import load_dotenv
import cohere
```

```python
load_dotenv()  # Load environment variables from .env file

try:
    # Using ClientV2 (the current standard)
    co = cohere.ClientV2()
    
    # Try to fetch models - if the key is bad, this will fail
    models = co.models.list()
    
    print("✅ Success! Your API key is valid and working.")
    print(f"You have access to {len(models.models)} models.")
    
    # Optional: Print the first 3 models you can use
    print("\nSample of available models:")
    for m in models.models[:3]:   # Adjust the number as needed
        print(f"- {m.name}")

except Exception as e:
    # This will catch 401 Unauthorized or other connection issues
    print(f"❌ Connection failed.")
    print(f"Error Details: {e}")
```

```python
# Check the response by sending a simple question to the model

user_question = "what is capital of India?"
    
response = co.chat(
    model="command-a-vision-07-2025",
    messages=[
        {
            "role": "user", 
            "content": user_question
        }
    ]
)

print(response.message.content[0].text)
```
